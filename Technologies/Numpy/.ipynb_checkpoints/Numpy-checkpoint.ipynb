## Integrals and Derivatives:

## Question 1:
Find the derivatives and draw it :$$ f(x)=1/x ^2*sin(x)$$

```python
x=np.linspace(1,10,100)
f_x=1/x**2 * np.sin(x)
y=f_x

plt.plot(x,y)

#Find Derivatives:
dydx= np.gradient(y, x)
plt.plot(x,dydx)

#Integrals:
y_int=np.cumsum(y) * (x[1]-x[0])
plt.plot(x,y_int)
```

## Question 2:

Let $$ y = e^{-x/10} \sin(x) $$. Consider 10,000 \( x \) intervals in the range \([0, 10]\)

1. Plot the function \( y \) vs. \( x \) in the range \([0, 10]\).
2. Compute the mean and standard deviation of \( y \) for \( x \) values in \([4, 7]\).
3. For \( x \) in the range \([4, 7]\), find the value \( y_m \) such that 80% of \( y \) values are less than \( y_m \).
4. Plot dy/dx vs x.
5. Find the locations where dy/dx=0.


```python
#Answer-1:
x=np.linspace(0,10,10000)
y=np.exp(-x/10)*np.sin(x)
plt.plot(x,y)
```

```python
#Answer 2:
y_update=y[(x>=4) * (x<=7)]
print("Mean:",np.mean(y_update))
print("Standard deviations",np.std(y_update))
```

```python
#Answer 3:
np.percentile(y[(x>=4) * (x<=7)],80)
```

```python
#Answer 4:
dydx=np.gradient(y,x)
plt.plot(x,dydx)
```

```python
#Answer 5:
x[1:][dydx[1:]*dydx[:-1]<0]
```

##  System of equations:
solve this:
$$3x+2y+z=4$$
$$4x-5y+4z=3$$
$$6x+z=0$$

```python
#Solution:
#Create matrix with cofficient:
A=np.array([
    [3,2,1],
    [4,-5,4],
    [6,0,1]
])
#Constrant part:
B=np.array([4,3,0])

#np.linalg.solve(A,B)
np.round(np.linalg.solve(A,B),3)
```

## Finding Eigenvalues:

```python
A=np.array([
    [4,2,2],
    [2,4,2],
    [2,2,4]
])

w,v=np.linalg.eig(A)
print("Eigen value:",w)
print("Eigen vector:",v)
```
