## Question-1:
$$Minimize f(x)=(x-3)^2$$


```python
#Answer:
def f(x):
    return (x-3)**2

res=minimize(f,2)
print("Minimum at:", res.x)
print("Minimum value:", res.fun)
print("Success?", res.success)
print("Message:", res.message)
print("Function evaluations:", res.nfev)
print("Iterations:", res.nit)

```

## Question 3 : Minimize a function with multiple variable and constrain:
Minimize this function:$$ f(x,y)=x^2+y^2$$
 Subject to the constraints:$$x+y=1$$
 $$x−y≥0$$
 $$x≥0.1$$
 Bounds:
 $$0≤x≤5,0≤y≤5$$
 


```python
#Answer:
def f(x):
    return x[0]**2 + x[1]**2
#Declear constrain:
cons=(
    {'type':'eq','fun':lambda x: x[0]+x[0]-1},
    {'type':'ineq','fun':lambda x: x[0]-x[0]},
     {'type':'ineq','fun':lambda x: x[0]-0.1}
)

#Bounds:
bounds=[(0,5),(0,5)]

#initial guess
guess=(0,0)
res = minimize(f, x0=guess, constraints=cons, bounds=bounds)

print("Minimum at:", res.x)
print("Minimum value:", res.fun)



 # Plot minimum point
plt.plot(res.x[0], res.x[1], 'ro') 
plt.grid()
```

## Question-4:
Interpolate this function:$$f(x)=sinx$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Known data
x = np.arange(0, 100, 10)        
y = np.sin(x)                     

# Interpolation function
f = interp1d(x, y, kind='cubic')

# Values where we want to interpolate
given_x = np.linspace(0, 90, 100)
out_y = f(given_x)                


# Plot original data points
plt.plot(x, y, 'o', label='Original Data (x, sin(x))')

# Plot interpolated line
plt.plot(given_x, out_y, '-', label='Linear Interpolation')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Interpolation using interp1d')
plt.legend()
plt.grid()
plt.show()
print("Original x values:", f.x)
print("Original y values:", f.y)

```

## Question-5: You have to find out the value of a,b using curve fitting from this function:$$f(x)=ax^2+b$$
And lastly draw curve

```python
#Answer:
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,10)
y=x*np.sin(x)

#GIven function
def f(x,a,b):
    return a*x**2+b
# pvt=stor vlaue of unkown variable
pvt,pcov=curve_fit(f,x,y)

a,b=pvt
print("Value of a:",a)
print("Value of a:",b)

print(f"Final equation f(x)={a:.2f}x^2+{b:.2f}" )
plt.plot(x,y,'o-',label='Orginal Curve')

f=a*x**2+b
plt.plot(x,f,'o-',label='Fitted curve Curve')
plt.title("Curve Fitting (Linear)")
plt.grid()
plt.legend()
```

## Question-6: You have to find The value of A,w and Phi from this preiodic function:
$$y=A⋅cos(ωt+ϕ)$$

```python
#Answer:
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,10,20)
y=np.linspace(0,10,20)*np.sin(t)


plt.title("Curve Fitting (Periodic motion)")
plt.plot(t,y,'o-',label="Orginal Curve")
plt.xlabel("T")
plt.xlabel("Y")
plt.legend()
plt.grid()


```

```python
#Curving fitting part:
def f(t,A,w,phi):
    return A*np.cos(w*t+phi)


initial_guess = [8, (2*np.pi)/4.20, 0]  # A, w, phi
pvt, pcov = curve_fit(f, t, y, p0=initial_guess)

# Plotting part:
A, w, phi = pvt
y_fit = f(t, A, w, phi)
print(f"Equation is : y={A:.2f}(cos({w:.2f}+{phi:.2f})")
plt.plot(t,y,'o-',label="Orginal Curve")
plt.plot(t,y_fit,'o-',label="Fitted Curve")
plt.title("Curve Fitting (Periodic motion)")
plt.xlabel("T")
plt.xlabel("Y")
plt.legend()
plt.grid()

```

```python
#A,W and Phi:
pvt
```

```python
#Find out error this value:
np.sqrt(np.diagonal(pcov))
```

## Question-7: Find derivative $f(x) = x^2 \cdot \sin(2x) \cdot e^{-x}$


```python
#import:
import numdifftools as nd

def f(x):
    return x**2 *np.sin(2*x) *np.exp(-x)
x=np.linspace(0,1,100)
#Just drow orginal curve
plt.plot(x,f(x),label='Orginal curve')


#Differtion:
df = nd.Derivative(f, step=1e-6)
#Draw differtiated curve:
plt.plot(x,df(x),label='Differntiated curve(1st order)')


df2 = nd.Derivative(df, step=1e-6,order=2)
plt.plot(x,df2(x),label='Differntiated curve(2nd order)',)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
```

## Question-9: Double integration:
solve it:
$$\int_{0}^{1} \int_{-x}^{x^2}\sin(x+y^2)dydx$$

```python
#Anser:
from scipy.integrate import dblquad
import numpy as np

# Define the function
f = lambda y, x: np.sin(x + y**2)  # Note: in dblquad, function takes (y, x

# Limits for y: from -x to x^2
lower_y = lambda x: -x
upper_y = lambda x: x**2

# Integrate over x from 0 to 1
res, err = dblquad(f, 0, 1, lower_y, upper_y)

print("Result:", res)
print("Error estimate:", err)
```

# Question-10:
solve this and findout the vlaue of  x


$$
\begin{bmatrix}
3 & 0 & 0 & 0 \\
2 & 1 & 0 & 0 \\
1 & 0 & 1 & 0 \\
1 & 1 & 1 & 1
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2 \\
x_3 \\
x_4
\end{bmatrix}
=
\begin{bmatrix}
4 \\
2 \\
4 \\
2
\end{bmatrix}
$$


```python
#Answer:
from scipy.linalg import solve_triangular
import numpy as np
A=np.array([
          [3,0,0,0],
          [2,1,0,0],
          [1,0,1,1],
          [1,1,1,1]
          ])
B=np.array([4,2, 4 ,2])
x=solve_triangular(A,B,lower=True)
#Solution:
print(x)
```
